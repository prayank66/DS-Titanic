```python
#From StakeOverFlow :How to display labels on top of each bar.

from matplotlib.pyplot import show
import seaborn as sns
sns.set(style="darkgrid")
titanic = sns.load_dataset("titanic")
total = float(len(titanic)) # one person per row 
#ax = sns.barplot(x="class", hue="who", data=titanic)
ax = sns.countplot(x="class", hue="who", data=titanic) # for Seaborn version 0.7 and more
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}'.format(height/total),
            ha="center") 
show()
```

```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'Child'
    else:
        return sex
```

```python
titanic_df['Person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
#dropping null values from Cabin
deck = titanic_df['Cabin'].dropna()
deck.head()
```

```python
levels=[]
for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']

cabin_df[0:10]
```

```python
# URL = matplotlib.org/users/colormaps.html 
# Winter is type of color 

sns.factorplot('Cabin',data=cabin_df,palette='winter_d',kind='count')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']
sns.factorplot('Cabin',data=cabin_df,palette='winter_d',kind='count')
```

```python
#from where people came from

sns.factorplot('Embarked',data=titanic_df,hue='Pclass',kind='count',order=['C','Q','S'])
```

```python
place = titanic_df['Embarked'].dropna()

def city(places):
    if places["Embarked"] == 'C':
        return "Cherbourg"
    elif places["Embarked"] == 'Q':
        return "Queenstown"
    else:
        return "Southampton"

    
titanic_df['City'] = titanic_df[['Embarked']].apply(city,axis=1)
```

```python
titanic_df[0:10]
```

```python
# Number of people boarded from different cities in different class
sns.factorplot('City',data=titanic_df,hue='Pclass',kind='count',order=['Cherbourg','Queenstown','Southampton'])
```

```python
# How many male,female and child boarded Titanic from different cities.
sns.factorplot('City',data=titanic_df,hue='Person',kind='count',order=['Cherbourg','Queenstown','Southampton'])
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
```

```python
# Did the deck have an effect on the passengers survival rate? Did this answer match up with your intuition?

deck = []

for i in titanic_df['Cabin']:
    
    if str(i) == 'NaN':
        deck.append(0)
    else:
        deck.append(str(i)[0])
        
titanic_df['Deck'] = deck

sns.factorplot('Deck',order=['A', 'B', 'C', 'D', 'E', 'F', 'G'], data=titanic_df, hue='Survived',kind='count')
```
